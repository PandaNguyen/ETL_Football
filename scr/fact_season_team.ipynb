```python
from pathlib import Path
```

```python
import soccerdata as sd
```

```python
import pandas as pd
```

```python
fbref = sd.FBref(leagues="ENG-Premier League",seasons=['2021','2122','2223','2324','2425'])

```

```python
# season_team_df = pd.DataFrame(season_team)
# season_team_df.head(5)
# season_team_df.to_csv("")
```

```python
player_season = fbref.read_player_season_stats()
player_season_df = pd.DataFrame(player_season)


out_path = Path("../data/fact_player_season_stats.csv")
out_path.parent.mkdir(parents=True, exist_ok=True)  
player_season_df.to_csv(out_path, index=False, encoding="utf-8")

player_season_df.to_csv("../data/fact_player_season_stats.csv")
```

```python
player_match = fbref.read_player_match_stats()
player_match_df = pd.DataFrame(player_match)


out_path = Path("../data/fact_player_match_stats.csv")
out_path.parent.mkdir(parents=True, exist_ok=True)  
player_match_df.to_csv(out_path, index=False, encoding="utf-8")

player_match_df.to_csv("../data/fact_player_match_stats.csv")
```
